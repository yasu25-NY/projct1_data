```python
import pandas as pd
```

```python
df = pd.read_csv('result_all.csv', encoding='utf-8')
```

```python
df['未着手'] = df['措置状況'].str.contains('未着手').astype(int)
df['完了'] = df['措置状況'].str.contains('完了').astype(int)
df['着手'] = ((df['措置状況'].str.contains('着手')) & (~df['措置状況'].str.contains('完了'))).astype(int)
```

```python
grouped = df.groupby(['都道府県名', '判定区分']).agg(
    総数 = ('トンネル名', 'count'),
    未着手数 = ('未着手', 'sum'),
    完了数 = ('完了', 'sum'),
    着手数 = ('着手', 'sum')
).reset_index()
```

```python
grouped['未着手率(%)'] = grouped['未着手数'] / grouped['総数'] * 100
grouped['完了率(%)'] = grouped['完了数'] / grouped['総数'] * 100
grouped['着手率(%)'] = grouped['着手数'] / grouped['総数'] * 100
```

```python
result = grouped[['都道府県名', '判定区分', '総数', '未着手数', '完了数', '着手数', '未着手率(%)', '完了率(%)', '着手率(%)']]
display(result)
```

```python
result.to_csv('PrefTotals.csv', index=False, encoding='utf-8-sig')
```
